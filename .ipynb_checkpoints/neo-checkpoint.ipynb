```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
from matplotlib import rc
import matplotlib.pyplot as plt
#한글로 편하게 하려고하는데 계속 한글이 깨져서 pit말고 sns 포기
plt.rcParams["font.family"] = "NanumGothic"
import datetime as dt
import seaborn as sns
import plotly.express as px
```

```python
#데이터 읽기 // dfc로 해놓고 중복제거하기 
dfc= pd.read_csv('C:/Users/김기석/Desktop/bigdata/neo.csv')
#잘 들어갔나 확인
dfc.shape
```

```python
#데이터 보기 상위10개
dfc.head(10)
```

```python
#중복된 데이터 값이 있는지 확인
is_vanza = dfc['id'] == 2513171
is_vanzaa = dfc[is_vanza]
is_vanzaa
```

```python
#id와 name 중복제거 // id와 name중복을 제거 하고 df에 넣기 ,  애초에 상대경로만 값이 다름(중복인상태서)
df=dfc.drop_duplicates(['id','name'])
```

```python
#지구한테 위험한지 안위험한지
df.hazardous.value_counts().plot(kind='bar')
plt.title('hazardous')
plt.show()
```
